```python
import pandas as pd
import numpy as np
```

```python
#Data 
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
```

```python
d = {"revenue": revenue, "expenses": expenses}
monthly_data = pd.DataFrame(d)
```

```python
def financial_analysys(df):
    df["profit"] = df.revenue - df.expenses
    df['pat'] = round((df.profit*0.7),2)
    df['profit margin'] = (df.pat/df.revenue*100).round(0).astype(str) + "%"
    df['type_of_month'] = df.pat.apply(lambda x: "Good Month" if x > df.pat.mean() else "Bad Month" )
    print("The Best Month is {}".format(df.pat.max()))
    print("The Worst Month is {}".format(df.pat.min()))
    return (df)
```

```python
financial_analysys(monthly_data)
```
